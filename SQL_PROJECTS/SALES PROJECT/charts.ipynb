```python
import matplotlib.pyplot as plt

# 1 Top 5 most selling products by quantity
top_5_product=["Wardrobe","Vegatables","Sofa","Dinning Table","Fruits"]
total_quantity_sold=[70,69,66,65,60]

plt.bar(top_5_product, total_quantity_sold, color="skyblue")
plt.title("Top 5 Most Selling Products by Quantity")
plt.ylabel("Quantity Sold")
plt.show()


# 2 Top 5 most frequently cancelled products
top_5_cancelled_product=["Comics","Sweater","Vegetables","Chaiir","T-shirt"]
total_cancelled=[24,23,21,21,20]

plt.bar(top_5_cancelled_product, total_cancelled, color="red")
plt.title("Top 5 Most Frequently Cancelled Products")
plt.ylabel("Number of Cancellations")
plt.show()


# 3 Peak purchase time of day
time_of_the_day=["EVENING","MORNING","NIGHT","AFTERNOON"]
total_orders=[514,512,496,474]

plt.plot(time_of_the_day, total_orders, marker="o")
plt.title("Peak Purchase Time of Day")
plt.ylabel("Total Orders")
plt.xlabel("Time of Day")
plt.show()


# 4 Top 5 highest spending customers in a year
coustomer_name=["Darshit Mann","Anahita Shenoy","Saira Ahluwalia","Gatik Khare","Samaira Subramaniam"]
total_spend=[507530,455637,447933,386156,357388]

plt.barh(coustomer_name, total_spend, color="green")
plt.title("Top 5 Highest Spending Customers ()")
plt.xlabel("Total Spend")
plt.show()


# 5 Revenue by product category (in crores)
product_category=["Accessories","Clothing","Books","Furniture","Electronics","Groceries"]
revenue_in_crores=[1.04,1.02,0.99,0.97,0.95,0.94]

plt.pie(revenue_in_crores, labels=product_category, autopct='%1.1f%%')
plt.title("Revenue by Product Category (in Crores)")
plt.show()


# 6 Cancellation / Return rate per category
product_category=["Books","Clothing","Electronics","Accessories","Furniture","Grocies"]
cancelled_percent=[25.98,25.71,24.76,23.55,22.83,22.06]
returned_percent=[25.68,24.58,20.85,31.50,23.41,23.56]

x = range(len(product_category))
plt.bar(x, cancelled_percent, width=0.4, label="Cancelled %")
plt.bar([i+0.4 for i in x], returned_percent, width=0.4, label="Returned %")
plt.xticks([i+0.2 for i in x], product_category)
plt.title("Cancellation / Return Rate per Category")
plt.ylabel("Percentage (%)")
plt.legend()
plt.show()


# 7 Preferred payment mode
paayment_mode=["Credit Card","EMI","Debit Card","Cash","UPI"]
total_count=[648,349,344,330,325]

plt.pie(total_count, labels=paayment_mode, autopct='%1.1f%%')
plt.title("Preferred Payment Mode")
plt.show()


# 8 Age group purchase behavior
age_group=["51+","36-50","26-35","18-25"]
total_purchase_in_crores=[1.44,1.94,1.37,1.16]

plt.bar(age_group, total_purchase_in_crores, color="purple")
plt.title("Age Group Purchase Behavior (in Crores)")
plt.ylabel("Total Purchases (Crores)")
plt.show()


# 9 Monthly sales trend
Month_Year=["2023-01","2023-02","2023-03","2023-04","2023-05","2023-06",
            "2023-07","2023-08","2023-09","2023-10","2023-11","2023-12","2024-01"]
total_sales_in_crores=[0.46,0.47,0.52,0.5,0.39,0.41,0.51,0.48,0.5,0.59,0.51,0.52,0.03]

plt.plot(Month_Year, total_sales_in_crores, marker="o", color="blue")
plt.title("Monthly Sales Trend (in Crores)")
plt.xlabel("Month-Year")
plt.ylabel("Sales (Crores)")
plt.xticks(rotation=45)
plt.show()


#  Gender vs Product Category
gender=["F","F","F","F","F","F","M","M","M","M","M","M"]
product_category=["Accessories","Books","Clothing","Electronics","Furniture" ,"Groceries",
                  "Accessories","Books","Clothing","Electronics","Furniture","Groceries"]
total_purchase=[156,151,180,147,183,165,171,180,174,160,163,166]

import pandas as pd
df = pd.DataFrame({"Gender": gender, "Category": product_category, "Purchase": total_purchase})
gender_category = df.pivot_table(index="Category", columns="Gender", values="Purchase", aggfunc="sum")

gender_category.plot(kind="bar")
plt.title("Gender vs Product Categories")
plt.ylabel("Number of Purchases")
plt.xlabel("Product Category")
plt.show()
```
